```python
#Assessment 2
# Make a Python program that calculates the grades based on the
# percentage obtained in four different subjects. Use the following
# grading scheme: 80% and above is A+, 70% to 79% is A, 60% to 69% is
# A-, 50% to 59% is B, 40% to 49% is B-, and below 40% is Fail


#created by SUJAL DEVKOTA
dsa=int(input("Enter the marks in DSA"))
java=int(input("Enter the marks in JAVA"))
cpp=int(input("Enter the marks in C++"))
python=int(input("Enter the marks in Python"))
total_marks= (dsa+java+cpp+python)
percentage=(total_marks/400)*100
if percentage>=80 and percentage<101:
    print('Your total percentage is ', percentage,'%','and you score A+')
    print('Excellent')
elif percentage>=70 and percentage<79:
      print('Your total percentage is ', percentage,'%','and you score A')
      print('Best')
elif percentage>=60 and percentage<69:
      print('Your total percentage is ', percentage,'%','and you score A-')
      print('Very Good')
elif percentage>=50 and percentage<59:
      print('Your total percentage is ', percentage,'%','and you score B')
      print('Good')
elif percentage>=40 and percentage <49:
      print('Your total percentage is ', percentage,'%','and you score B-')
      print('Average')
elif percentage<=48:
      print('Your total percentage is ', percentage,'%','and you get fail in exam')
      print('Poor')
else:
    print('Marks entry error')
```

```python
#Assessment 3

# Write a Python program that generates a random number between 1
# and 100 and asks the user to guess the number. The program should
# provide feedback to the user (e.g., "Too high" or "Too low") based on
# their guess and count the number of attempts it takes to guess the
# correct number. Once the user guesses the correct number, the
# program should display the number of attempts.

#created by SUJAL DEVKOTA
import random

randnum = random.randint(1, 100)
print(randnum) 
guessCount = 0

while True:
    userGuess = int(input('Guess the number between 1 and 100: '))
    guessCount += 1

    if userGuess < randnum:
        print("Your guess is low")
    elif userGuess > randnum:
        print("Your guess is high")
    else:
        print("Congratulations! Your guess is correct.")
        break

print("You guessed the number in", guessCount, "attempts.")


```
